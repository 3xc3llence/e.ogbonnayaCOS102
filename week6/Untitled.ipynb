```python
from tkinter import *
from tkinter import Tk, Label, Entry, Button, StringVar, IntVar, Toplevel

# Define menu items and prices 
menu_items = {
    "Jollof rice": 350,
    "Coconut fried rice": 350,
    "Jollof Spaghetti": 400,
    "Savory Beans": 350,
    "Roasted Sweet Potatoess": 300,
    "Fried Plantains": 150,
    "Boiled Yam": 150,
    "Sweet Chilli  Chicken": 1100,
    "Grilled Chicken Wings": 400,
    "Fried Beef": 500,
    "Boiled Egg": 200,
    "Sauteed Sausage": 200,
    "Eba": 100,
    "Poundo Yam": 100,
    "Semo": 100,
    "Atama Soup": 450,
    "Egusi Soup": 480,
    "Water": 200,
    "Glass Drink(35cl)": 150,
    "PET Drink(35cl)": 300,
    "PET Drink(50cl)": 350,
    "Glass/Canned Malt": 500,
    "Fresh Yo": 600,
    "Pineapplee Juice": 350,
    "Mango Juice": 350,
    "Zobo Drink": 350
    
}

# Function to calculate total before discount
def calculate_total():
    global total_price
    total_price = 0
    for item, quantity in order_dict.items():
        total_price += menu_items[item] * quantity
    global final_price
    discount = 0
    if total_price < 1000:
        discount = 0
    elif total_price < 2500:
        discount = 0.1
    elif total_price < 5000:
        discount = 0.15
    else:
        discount = 0.25
    final_price = total_price * (1 - discount)
    discount_label.config(text=f"Discount: N{total_price * discount:.2f}")
    final_label.config(text=f"Final Price: N{final_price:.2f}")


 
# Function to show order details in a new window
def show_order():
    order_window = Toplevel(root)
    order_window.title("Order Details")
    order_text = "Order Summary:\n"
    for item, quantity in order_dict.items():
        order_text += f"{item} x {quantity} - N{menu_items[item] * quantity:.2f}\n"
    order_label = Label(order_window, text=order_text)
    order_label.pack()

# Function to add items to order
def add_to_order():
    selected_item = item_var.get()
    quantity = int(quantity_entry.get())
    if selected_item in menu_items:
        if selected_item in order_dict:
            order_dict[selected_item] += quantity
        else:
            order_dict[selected_item] = quantity
        quantity_entry.delete(0, 'end')
        item_var.set("")
    else:
        print("Invalid item selection")


root = Tk()
root.title("PAU CAFETERIA MENU")
root.geometry("800x600")
# Customer name label and entry
customer_label = Label(root, text="Welcome to Pau Cafeteria Menu, please pick what you will like to eat")
customer_label.pack()
customer_label = Label(root, text="Customer Name:")
customer_label.pack()
customer_name = StringVar()
customer_entry = Entry(root, textvariable=customer_name)
customer_entry.pack()

# Order list (initially empty)
order_dict = {}

# Menu selection and quantity
item_label = Label(root, text="Select Item:")
item_label.pack()
item_var = StringVar()
item_dropdown = OptionMenu(root, item_var, *menu_items.keys())
item_dropdown.pack()

quantity_label = Label(root, text="Quantity:")
quantity_label.pack()
quantity_entry = Entry(root)
quantity_entry.pack()

add_button = Button(root, text="Add to Order", command=add_to_order)
add_button.pack()

# Button to calculate total and discount
calculate_button = Button(root, text="Calculate Total", command=calculate_total)
calculate_button.pack()

# Labels to display total before discount, discount amount, and final price
total_label = Label(root)
total_label.pack()
discount_label = Label(root)
discount_label.pack()
final_label = Label(root)
final_label.pack()

# Button to show order details
show_order_button = Button(root, text="Show Order", command=show_order)
show_order_button.pack(side=BOTTOM)
   
root.mainloop()
```
